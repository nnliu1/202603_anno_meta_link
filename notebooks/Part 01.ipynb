# FAIR Principles and the RDMO Framework

## 1. Theory: The FAIR Principles & RDMO

The **FAIR Principles** are not just abstract ideas; they are actionable requirements. In this training, we use **RDMO** to bridge the gap between "planning" and "doing."

### Objective: Mapping FAIR to RDMO Categories

RDMO organises a Data Management Plan (DMP) into question categories. Here is how they align with FAIR:

| **FAIR Principle** | **RDMO Question Category** | **Key Focus**                                        |
| ------------------ | -------------------------- | ---------------------------------------------------- |
| **F**indable       | Metadata & PIDs            | Can a machine find this via a DOI?                   |
| **A**ccessible     | Data Access & Storage      | Who can get the data? (Open vs. Embargo)             |
| **I**nteroperable  | Standards & Formats        | Is the data in a non-proprietary format (CSV, JSON)? |
| **R**eusable       | Licenses & Documentation   | Can someone else legally and technically use it?     |

## 3. Theory: The 'F' in Findable – PIDs and Metadata
Addressing the **F** during the collection phase is critical. If you don't define your metadata early, the data becomes a "dark data" silo that no one can search for.

### Persistent Identifiers (PIDs)

- **DOI (Digital Object Identifier):** Best for datasets and publications.
- **Handle:** Flexible, often used in institutional repositories.
- **ORCID:** For identifying the researchers themselves.

## 4. Practical: Defining a Metadata Standard with LinkML
**Theory:** To ensure **Interoperability (I)**, we often choose a discipline-specific standard (like **OEMEetadata** for energy) over a general one (like Dublin Core).

In this exercise, we will use **LinkML** to "formalise" a metadata requirement that would typically be a question in RDMO.

```python
import json
from linkml.validator import validate

# We are validating using a schema that enforces a PID (DOI) and a License
with open('../data/title_doi_license.json', 'r') as f:
    test_data = json.load(f)

# Test data: One valid, one invalid (wrong DOI format)
# test_data = {
#     "title": "Solar Output 2026",
#     "pid_doi": "10.1234/zenodo.5678",
#     "license": "CC-BY-4.0"
# }

for record in test_data:
    print(f"Validating: {record.get('title', 'Unknown Title')}")

    # Validate each individual record in the list
    report = validate(record, "../schema/fair_schema.yaml", "DatasetMetadata")

    if not report.results:
        print("✅ Validation Success!")
    else:
        print(f"❌ Errors found: {report.results}")
    print("-" * 20)
```
